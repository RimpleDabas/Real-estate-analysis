```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

from scipy.stats import linregress
# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#read initial files
rental_df = pd.read_csv("Average Rents.csv")
units_df = pd.read_csv("Number of Units.csv")
vacancy_df = pd.read_csv("Vacancy Rates.csv")
```

```python
# display 
rental_df.head()
```

```python
Provinces = rental_df["Province"].unique()
print(Provinces)
Province_ids_number = np.arange(1,len(Provinces)+1 )
print(len(Province_ids_number))
Province_ids = ['Province' + str(each) for each in Province_ids_number]
print(Province_ids)
Province_df = pd.DataFrame({"Province_id" : Province_ids , 'Provinces' : Provinces})
Province_df
Province_df.to_csv("Provinces.csv", index=False)
```

```python
Neighbourhood = rental_df["Neighbourhood"].unique()
#print(Provinces)
ids_number = np.arange(1,len(Neighbourhood)+1 )
#print(len(Province_ids_number))
ids = ['Neighbourhood' + str(each) for each in ids_number]
#print(Province_ids)
Neighbourhood_df = pd.DataFrame({"Neighbourhood_id" : ids , 'Neighbourhood' :Neighbourhood})

Neighbourhood_df.to_csv("Neighbourhood.csv", index=False)
```

```python
Dwelling = rental_df["Dwelling Type "].unique()
#print(Provinces)
ids_number = np.arange(1,len(Dwelling)+1 )
#print(len(Province_ids_number))
ids = ['Dwelling' + str(each) for each in ids_number]
#print(Province_ids)
Dwelling_df = pd.DataFrame({"Dwelling_id" : ids , 'Dwelling Type' : Dwelling})

Dwelling_df.to_csv("Dwelling.csv", index=False)
```
